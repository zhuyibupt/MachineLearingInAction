```python
#-*- coding: utf-8 -*-
import KNN
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
datingDataMat, datingLabels = KNN.file2matrix('datingTestSet.txt')
datingDataMat
```

```python
datingLabels[:20]
```

```python
fig = plt.figure(num = 1, figsize = (12,8))
plt.scatter(datingDataMat[:, 1], datingDataMat[:, 2])
plt.xlabel("percentage of playing vedio game")
plt.ylabel("liters of ice cream per week")
plt.show()
```

```python
fig = plt.figure(num = 2, figsize = (18,12))
ax1 = fig.add_subplot(221)
ax1 = plt.scatter(datingDataMat[:, 1], datingDataMat[:, 2], 15.0, np.array(datingLabels)) # sizecolor
plt.xlabel("percentage of time spent playing vedio game")
plt.ylabel("liters of ice cream consumed per week")

ax2 = fig.add_subplot(222)
ax2 = plt.scatter(datingDataMat[:, 0], datingDataMat[:, 1], 15.0, np.array(datingLabels)) # sizecolor
plt.xlabel("frequent flier miles earned per year")
plt.ylabel("percentage of time spent playing vedio game")

ax3 = fig.add_subplot(223)
ax3 = plt.scatter(datingDataMat[:, 0], datingDataMat[:, 2], 15.0, np.array(datingLabels)) # sizecolor
plt.xlabel("frequent flier miles earned per year")
plt.ylabel("liters of ice cream consumed per week")
plt.show()
```

```python
fig = plt.figure(num = 3, figsize = (12,8))
ax = fig.add_subplot(111)
x_1, x_2, x_3, y_1, y_2, y_3 = [],[],[],[],[],[]
for i in range(len(datingLabels)):
    if datingLabels[i] == 1:
        x_1.append(datingDataMat[i][0])
        y_1.append(datingDataMat[i][1])
    if datingLabels[i] == 2:
        x_2.append(datingDataMat[i][0])
        y_2.append(datingDataMat[i][1])
    if datingLabels[i] == 3:
        x_3.append(datingDataMat[i][0])
        y_3.append(datingDataMat[i][1])
type1 = plt.scatter(x_1, y_1, 45, 'orange', alpha = 0.9)
type2 = plt.scatter(x_2, y_2, 45, 'green', alpha = 0.6)
type3 = plt.scatter(x_3, y_3, 45, 'purple', alpha = 0.7)
plt.xlabel("mileage of flight per year")
plt.ylabel("percentage of time spent playing vedio game")
plt.legend((type1, type2, type3), ("didntLike", "smallDoses", "largeDoses"))
plt.show()
```
